```python
# convert embeddings to numpy arrays for use with FAISS
embeddings_dataset = retr_dataset["train"].map(
    lambda x: {"embeddings": model.encode(x["passage"])},
    batched=True,
    batch_size=1024
)

embeddings_dataset
```

```python
# create faiss index for the embeddings dataset
import faiss

embeddings_dataset.add_faiss_index(column="embeddings", metric_type=faiss.METRIC_INNER_PRODUCT)
```

### **Generate Negatives**

```python
import random
import numpy as np

def sample_hard_negatives(query, query_id, k=150, cut=20, num_negatives=5):
  query_embedding = model.encode(query, show_progress_bar=False)
  query_embedding.shape

  scores, samples = embeddings_dataset.get_nearest_examples(
      "embeddings", query_embedding, k=k
  )
  # print("Top Scores:", scores)
  score_cutoff = 0.6

  for i, score in enumerate(scores):
    if score < score_cutoff:
      break
  cut = min(50, max(i, cut))

  # print("Cut:", cut)
  # if query_id in samples["query_id"]:
  #   # take passages ranked below relevant doc, but take docs ranked after 40 if the relevant doc has rank below 40
  #   q_idx = min(samples["query_id"].index(query_id)+1, 15)
  #   cut = max(q_idx, cut)

  # if cut > 30:
  #   print("CUT:", cut, (query, query_id))

  offset = 100
  candidate_scores, candidate_qids, candidate_pass_ids, candidate_passages = scores[cut:cut+offset], samples["query_id"][cut:cut+offset], samples["passage_id"][cut:cut+offset], samples["passage"][cut:cut+offset]
  negatives, neg_pass_ids = [], set()

  while len(negatives) < num_negatives:
    idx = random.randint(0, len(candidate_passages)-1)
    q_id, pass_id = candidate_qids[idx], candidate_pass_ids[idx]
    passage = candidate_passages[idx]
    if q_id != query_id and pass_id not in neg_pass_ids:
      neg_pass_ids.add(pass_id)
      negatives.append({
          "passage_id": pass_id,
          "passage": passage,
          "cosine_sim_score": candidate_scores[idx]
        })

  return sorted(negatives, key=lambda x: x["cosine_sim_score"], reverse=True)
```

```python
idx = random.randint(0, 1000)
Q, QID = embeddings_dataset[idx]["query"], embeddings_dataset[idx]["query_id"]
Q, QID
```

```python
sample_hard_negatives(Q, QID, num_negatives=5)
```

```python
rel_dataset_with_negatives = retr_dataset.map(lambda row: {"negative_passages": sample_hard_negatives(row["query"], row["query_id"], num_negatives=16)})
rel_dataset_with_negatives
```
