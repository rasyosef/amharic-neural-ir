#### **Create and Prepare embedding dataset**

```python
from datasets import load_dataset

dataset = load_dataset("rasyosef/Amharic-Passage-Retrieval-Dataset-V2")
dataset
```

```python
from datasets import concatenate_datasets

# load test dataset
test_dataset = dataset["test"]
train_dataset = dataset["train"]
corpus_dataset = concatenate_datasets([train_dataset, test_dataset])

corpus_dataset
```

#### **Create baseline and evaluate pretrained model**

```python
# Convert the datasets to dictionaries
corpus = dict(
    zip(corpus_dataset["passage_id"], corpus_dataset["passage"])
) # Our corpus (cid => document)
queries = dict(
    zip(test_dataset["query_id"], test_dataset["query"])
) # Our queries (qid => question)
```

```python
relevant_docs = {}
for row in test_dataset:
  relevant_docs[row["query_id"]] = [row["passage_id"]]
```

```python
len(queries), len(corpus), len(relevant_docs)
```

```python
# convert embeddings to numpy arrays for use with FAISS
from datasets import Dataset

corpus_dedup_dataset = Dataset.from_dict({
    "passage_id": list(corpus.keys()),
    "passage": list(corpus.values()),
})

embeddings_dataset = corpus_dedup_dataset.map(
    lambda x: {"embeddings": model.encode(x["passage"])},
    batched=True,
    batch_size=256
)

embeddings_dataset
```

#### **FAISS**


```python
# create faiss index for the embeddings dataset
import faiss

embeddings_dataset.add_faiss_index(
    column="embeddings", 
    metric_type=faiss.METRIC_INNER_PRODUCT
)
```

```python
from sentence_transformers import CrossEncoder

reranker_id = "rasyosef/RoBERTa-Amharic-Reranker-Base"
reranker = CrossEncoder(
    reranker_id, 
    device="cuda" if torch.cuda.is_available() else "cpu"
  )
```

```python
def retrieve_and_rerank(query, k=100, rerank_k=10):
  query_embedding = model.encode(query)
  query_embedding.shape

  scores, samples = embeddings_dataset.get_nearest_examples(
      "embeddings", query_embedding, k=k
  )

  retrieved_docs = [{"passage": txt, "corpus_id": cid} for txt, cid in zip(samples['passage'], samples['passage_id'])]
  rerank_scores = reranker.predict([[query, doc["passage"]] for doc in retrieved_docs])
  for i in range(len(rerank_scores)):
    retrieved_docs[i]["score"] = rerank_scores[i]

  return sorted(retrieved_docs, key=lambda x: x["score"], reverse=True)[:rerank_k]

query = "·àÖ·ãù·â£·ãä ·àò·ãã·åÆ"
retrieve_and_rerank(query, k=5, rerank_k=3)

import numpy as np

def retrieve(query, k=100):
  query_embedding = model.encode(query)
  query_embedding.shape

  scores, samples = embeddings_dataset.get_nearest_examples(
      "embeddings", query_embedding, k=k
  )

  retrieved_docs = [
      {"passage": txt, "corpus_id": cid, "score": score.item()} 
      for txt, cid, score in zip(samples['passage'], samples['passage_id'], scores)
    ]

  return sorted(retrieved_docs, key=lambda x: x["score"], reverse=True)[:k]

retrieve_and_rerank(query, k=5, rerank_k=3)
retrieve(query, k=5)
```

```python
from sentence_transformers.evaluation import InformationRetrievalEvaluator

ir_evaluator = InformationRetrievalEvaluator(
      queries=queries,
      corpus=corpus,
      relevant_docs=relevant_docs,
      mrr_at_k=[10],
      ndcg_at_k=[10],
      precision_recall_at_k=[5, 10],
  )
```

### Retrieve

```python
from tqdm import tqdm

retrieved = [None]*len(queries)
for query_idx, query in enumerate(tqdm(list(queries.values()))):
  retrieved[query_idx] = retrieve(query, k=10)

retrieved[0], retrieved[4]
```

```python
ir_evaluator = InformationRetrievalEvaluator(
      queries=queries,
      corpus=corpus,
      relevant_docs=relevant_docs,
      mrr_at_k=[10],
      ndcg_at_k=[10],
      precision_recall_at_k=[5, 10],
  )
```

```python
# Evaluate the model
results = ir_evaluator.compute_metrics(retrieved)

# dim = EMBED_DIM
metrics = [
    "recall@k",
    "mrr@k",
    "ndcg@k",
  ]

for metric in metrics:
  for k in results[metric]:
    print(f"{metric.split('@')[0]}@{k}: {round(results[metric][k], 3)}")
```
